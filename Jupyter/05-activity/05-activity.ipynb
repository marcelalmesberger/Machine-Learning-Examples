```python
# Activity prediction
# Based on: https://beckernick.github.io/activity-prediction/
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_features = pd.read_csv('features.txt',header=None)
```

```python
df_activity_labels = pd.read_csv('activity_labels.txt', sep='\s+',header=None)
```

```python
X_train = pd.read_csv('X_train.txt',delim_whitespace=True,header=None)
```

```python
y_train = pd.read_csv('y_train.txt',delim_whitespace=True,header=None)
```

```python
X_test = pd.read_csv('X_test.txt',delim_whitespace=True,header=None)
```

```python
y_test = pd.read_csv('y_test.txt',delim_whitespace=True,header=None)
```

```python
from sklearn.svm import LinearSVC
```

```python
clf_svc = LinearSVC(random_state = 12, C=0.1, max_iter=5000)
```

```python
clf_svc.fit(X_train, y_train.values.flatten())
```

```python
clf_svc.score(X_test, y_test)
```

```python
crosstab = pd.crosstab(y_test.values.flatten(), clf_svc.predict(X_test),
                          rownames=['True'], colnames=['Predicted'],
                          margins=True)
crosstab
```

```python
# Remove last column and row
crosstab_clean = crosstab.iloc[:-1, :-1]
crosstab_clean.columns = df_activity_labels.values
crosstab_clean.index = df_activity_labels.values
crosstab_clean
```
