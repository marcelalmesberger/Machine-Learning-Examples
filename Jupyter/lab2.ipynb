```python
df1 = pd.DataFrame(np.random.randn(25).reshape(5,5))
df1.loc[2:3,1:2] = np.nan
df1
```

```python
df_filled = df1.fillna(0)
df_filled
```

```python
default_values = {1:-0.5,2:0.5}
df_default = df1.fillna(value = default_values)
df_default
```

```python
df_previous = df1.fillna(method="ffill")
df_previous
```

```python
means = df1.mean(skipna=True)
df_means = df1.fillna(value={1:means[1],2:means[2]})
df_means
```
